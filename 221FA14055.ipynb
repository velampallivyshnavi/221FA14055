```python
df=pd.read_csv("/content/NYC Taxi Data.csv")
```

```python
# prompt: from the above code select two columns as 	pickup_longitude and dropoff_longitude and generate linear regression graph  in python

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Select the columns
X = df[['pickup_longitude']]
y = df['dropoff_longitude']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, predictions, color='red')
plt.xlabel('pickup_longitude')
plt.ylabel('dropoff_longitude')
plt.title('Linear Regression')
plt.show()
```

```python
# prompt: from the above code select two columns as 	pickup_longitude and dropoff_longitude and generate linear regression graph  in python by using path

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
df = pd.read_csv("/content/NYC Taxi Data.csv")

# Select the columns
X = df[['pickup_longitude']]
y = df['dropoff_longitude']

# Handle missing values by dropping rows with NaN
df = df.dropna(subset=['pickup_longitude', 'dropoff_longitude'])

# Select the columns again after handling missing values
X = df[['pickup_longitude']]
y = df['dropoff_longitude']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, predictions, color='red')
plt.xlabel('pickup_longitude')
plt.ylabel('dropoff_longitude')
plt.title('Linear Regression')
plt.show()
```
